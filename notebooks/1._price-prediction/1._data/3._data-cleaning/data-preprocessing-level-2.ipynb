```python
indices = df[df['sector'] == 'new'].index
```

```python
def update_new_sector(row: pd.Series) -> pd.Series:
    if row['sector'] == 'new':
        if row['society'] == "takshila heights sector 37 c":
            row['sector'] = 'sector 37'
        elif row['society'] == 'dlf 76':
            row['sector'] = 'sector 76'
        elif row['society'] == 'green court':
            row['sector'] = 'sector 90'
        elif row['society'] == 'sare homes':
            row['sector'] = 'sector 92'
    else:
        pass
    return row

df = df.apply(update_new_sector, axis=1)
```
